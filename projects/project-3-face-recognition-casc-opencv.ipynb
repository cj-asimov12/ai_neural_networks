```python
import sys
import matplotlib.pyplot as plt
imagePath = "C:\\Users\\black\\Desktop\\ai_py\\code_files\\p2\\tom_cruise.jpg"
cascPath = "C:\\Users\\black\\Desktop\\ai_py\\code_files\\p2\\haarcascade_frontalface_default.xml"
```

## Loading face cascade (haar) 

```python
faceCascade = cv2.CascadeClassifier(cascPath)
import os
for dirname, _, filenames in os.walk('C:\\Users\\black\\Desktop\\ai_py\\code_files\\p2\\'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

## Reading the image file 

```python
# Read the image
image = cv2.imread(imagePath)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

## Performing face recognition 

```python
# Detect faces in the image
faces = faceCascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(30, 30),
    #flags = cv2.cv.CV_HAAR_SCALE_IMAGE
    flags = cv2.CASCADE_SCALE_IMAGE
)
```

## Evaluating results 

```python
print ("Found {0} faces!".format(len(faces)))

# Draw a rectangle around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

## Viewing detected faces 

```python
#cv2.imshow("Faces found", image)

plt.imshow(image, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```
